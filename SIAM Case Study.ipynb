```python
# Importing pandas 
import pandas as pd
```

```python
# Reading the dataset .
# skiprows = 3 means to skip first 3 rows 
# header = [1,2,3] means after skipping rows make 1st, 2nd & 3rd row as header. Now it is Multi-row header.
# skipfooter = 1 means to skip last row since excel has report published date.
# na_values = 'NA' means that to consider NA has NaN values.
```

# Combining Dataset for CV

```python
# Here CV represents Commerical Vehicles .
# For Commerical Vehicles , data is present quaterly i.e in month of June, September, December and March. 
```

```python
# Reading the dataset .
# There are 2 sheets in same excel file . So we need to use this option to select required sheet_name
# skiprows = 3 means to skip first 3 rows 
# header = [1,2,3] means after skipping rows make 1st, 2nd & 3rd row as header. Now it is Multi-row header.
# skipfooter = 1 means to skip last row since excel has report published date.
# na_values = 'NA' means that to consider NA has NaN values.
```

```python
df_jun_cv = pd.read_excel('2022_6.xlsx',sheet_name ='Report-IV-CVs',skiprows=3,header=[1,2,3],skipfooter=1,na_values='NA',)
df_jun_cv
```

```python
df_jun_cv_dropna = df_jun_cv.dropna(axis=0,thresh=6)
df_jun_cv_drop_dup = df_jun_cv_dropna.drop_duplicates(subset = [(      'Category', 'Segment/Subsegment', 'Manufacturer')],ignore_index=True)
df_jun_cv_reset = df_jun_cv_drop_dup.reset_index(drop=True)
df_jun_cv_reset
```

```python
# Same steps are followed for Sept 2022 for Commerical Vehicles
```

```python
df_sept_cv = pd.read_excel('2022_9.xlsx',sheet_name ='Report-IV-CVs',skiprows=3,header=[1,2,3],skipfooter=1,na_values='NA',)
df_sept_cv_dropna = df_sept_cv.dropna(axis=0,thresh=6)
df_sept_cv_drop_dup = df_sept_cv_dropna.drop_duplicates(subset = [(      'Category', 'Segment/Subsegment', 'Manufacturer')],ignore_index=True)
df_sept_cv_reset = df_sept_cv_drop_dup.reset_index(drop=True)
df_sept_cv_reset
```

```python
# Same steps are followed for Dec 2022 for Commerical Vehicles
```

```python
df_dec_cv = pd.read_excel('2022_12.xlsx',sheet_name ='Report-IV-CVs',skiprows=3,header=[1,2,3],skipfooter=1,na_values='NA',)
df_dec_cv_dropna = df_dec_cv.dropna(axis=0,thresh=6)
df_dec_cv_drop_dup = df_dec_cv_dropna.drop_duplicates(subset = [(      'Category', 'Segment/Subsegment', 'Manufacturer')],ignore_index=True)
df_dec_cv_reset = df_dec_cv_drop_dup.reset_index(drop=True)
df_dec_cv_reset
```

```python
# Same steps followed for March 2023 for Commerical Vehicles
```

```python
df_mar_cv = pd.read_excel('2023_3.xlsx',sheet_name ='Report-IV-CVs',skiprows=3,header=[1,2,3],skipfooter=1,na_values='NA',)
df_mar_cv_dropna = df_mar_cv.dropna(axis=0,thresh=6)
df_mar_cv_drop_dup = df_mar_cv_dropna.drop_duplicates(subset = [(      'Category', 'Segment/Subsegment', 'Manufacturer')],ignore_index=True)
df_mar_cv_reset = df_mar_cv_drop_dup.reset_index(drop=True)
df_mar_cv_reset
```

```python
# Merging Jun 2022 & Sept 2022 for Commerical Vehicles
```

```python
dfjundfsept = pd.merge(df_jun_cv_reset,df_sept_cv_reset,how='outer')
dfjundfsept
```

```python
# Merging previous created dataframe with Dec 2022 for Commerical Vehicles
```

```python
dfjundfdec = pd.merge(dfjundfsept,df_dec_cv_reset,how='outer')
dfjundfdec
```

```python
# Merging previous created dataframe with March 2023
```

```python
dfjundfmar = pd.merge(dfjundfdec,df_mar_cv_reset,how='outer')
dfjundfmar
```

```python
# Merged excel files for Commerical Vehicles.
```

```python
dfjundfmar.to_excel('CV_all_files.xlsx')
```
